```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import quandl
import pickle
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
# get btc from quandl (source: https://blog.patricktriest.com/analyzing-cryptocurrencies-python/)
def get_quandl_data(quandl_id):
    '''Download and cache Quandl dataseries'''
    cache_path = '{}.pkl'.format(quandl_id).replace('/','-')
    try:
        f = open(cache_path, 'rb')
        df = pickle.load(f)   
        print('Loaded {} from cache'.format(quandl_id))
    except (OSError, IOError) as e:
        print('Downloading {} from Quandl'.format(quandl_id))
        df = quandl.get(quandl_id, returns="pandas")
        df.to_pickle(cache_path)
        print('Cached {} at {}'.format(quandl_id, cache_path))
    return df

btc_eur_price_kraken = get_quandl_data('BCHARTS/KRAKENEUR')
btc_eur_price_kraken
```

```python
btc = btc_eur_price_kraken
# playing with the plotting
btc[['Close','High','Low']].plot(grid= True, figsize=(8,5))
```

```python
btc = btc.replace(0, np.nan).fillna(method='ffill')


### creating a trendline 42d and 252d
btc['42d'] = np.round(btc['Close'].rolling(window=42, center= False).mean(),2)
btc['252d'] = np.round(btc['Close'].rolling(window=252, center= False).mean(),2)
# showing the max minus the min to say variance per day (my idea ;)
btc['day_range'] = btc['High']-btc['Low']
# plotting upward and downward trends
btc[['Close','42d','252d', 'day_range']].plot(grid= True, figsize=(8,5))
# ANALYSIS FROM THE BOOK
# the 42d trend is above the 252d trend =  buy (go long)
# the 42d trend is within a range of +/-SD points around the 252d trend =  wait (park in cash)
# the 42d trend is below the 252d trend =  sell (go short)
```

```python
# Calculate the regime (the connection between the 42d trend and the 25d trend)
btc['42-252'] = btc['42d']-btc['252d']
SD = 50
btc['Regime'] = np.where(btc['42-252'] > SD, 1, 0)
btc['Regime'] = np.where(btc['42-252'] < -SD, -1, btc['Regime'])
regime_count = btc['Regime'].value_counts()
print("On", regime_count[1],"dates the 42 trend lies more than",SD,"SD points above the 252d trend",
      "On", regime_count[0],"dates the 42 trend is +- ",SD,"SD points identical to the 252d trend",
      "On", regime_count[-1],"dates the 42 trend lies more than",SD,"SD points below the 252d trend")
regime_count
btc['Regime'].plot(lw=1.5)
```

```python
btc['Market'] = np.log(btc['Close']  / btc['Close'].shift(1) )
btc['Strategy'] = btc['Regime'].shift(1) * btc['Market']
btc[['Market','Strategy']].cumsum().apply(np.exp).plot(grid=True, figsize=(8,5))
```
